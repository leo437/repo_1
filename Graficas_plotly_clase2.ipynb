# Gráfica de funciones interactivas con la librería plotly.graph_objects
## Graficar la función $y=xcos(x^2)$ en el intervalo $[-3,3]$

```python
import numpy as np
import plotly.graph_objects as go

# Crear los valores de x en el intervalo [-3,3]
x=np.linspace(-3,3,200)

# Crear los valores de y
y=x*np.cos(1.5*x**2)

# Crear la figura
fig=go.Figure()

# Agregar la gráfica de la función
fig.add_trace(go.Scatter(x=x,y=y,mode='lines',name='y=xcos(x^2)'))
fig.show()
```

```python
# Resolver x^2=2x+8
x=np.linspace(-4,5,100)
y1=x**2
y2=2*x+8


fig=go.Figure()
fig.add_trace(go.Scatter(x=x,y=y1,mode='lines',name='y1=f(x)'))
fig.add_trace(go.Scatter(x=x,y=y2,mode='lines',name='y2=g(x)'))
fig.show()
```
